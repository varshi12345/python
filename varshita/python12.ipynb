```python
contact={}
def addcontact(name,phone):
    if name not in contact:
        contact[name]=phone
        print("contact %s added"%name)
    else:
        print("contact %s already exists"%name)
    return
addcontact('Anil',9000090000)
addcontact('Harsha',9500500044)
addcontact('Kranth',9800890012)
print(contact)
```
